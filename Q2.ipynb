```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
img=cv2.imread(path+'pic.jpg')
plt.imshow(img)
```

```python
img_bgr = cv2.imread(path+"pic.jpg",cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
```

```python
imgBlur = cv2.GaussianBlur(img_gray,(11,11),0)
plt.imshow(imgBlur, cmap='gray')
```

```python
imgCanny = cv2.Canny(imgBlur,30,250)
plt.imshow(imgCanny, cmap='gray')
```

```python
(cnt, hierarchy) = cv2.findContours(imgCanny.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
cv2.drawContours(img_rgb, cnt, -1, (0,250,0), 2)
plt.imshow(img_rgb)
```
